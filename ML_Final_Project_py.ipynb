```python
# Print Performance

def performance(output, gd):
  from sklearn.metrics import accuracy_score
  print("Accuracy ", round(accuracy_score(output, gd), 3))

  from sklearn.metrics import confusion_matrix
  cnf = confusion_matrix(output, gd)
  tp = cnf[0][0]
  fp = cnf[0][1]
  tn = cnf[1][1]
  fn = cnf[1][0]
  print("Sensitivity= ", round(tp/(tp+fn), 3))
  print("Specificity= ", round(tn/(tn+fp), 3))
  print("Precision= ", round(tp/(tp+fp), 3))
  print("Negative Predictive Value= ", round(tn/(tn+fn), 3))

def roc(y, output):
  from sklearn.metrics import roc_curve
  from sklearn.metrics import roc_auc_score
  auc = roc_auc_score(y, output)
  print(f'AUC: {auc :.4f}')
  import matplotlib.pyplot as plt
  fpr, tpr, thresholds = roc_curve(y, output, pos_label=1)
  plt.plot(fpr, tpr, color='red',label='ROC curve (area = %0.2f)' )
  plt.grid(color='grey', linewidth=0.5)
  plt.plot([0, 1], [0, 1], color='green', linestyle='--')
  plt.show()
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def softmax(x):
  return np.exp(x) / np.sum(np.exp(x))

def loss_f(x, y, w):
  x_loss = []
  for i in range(0, len(y)):
    x_loss.append(sigmoid( -y[i,:].dot( w.T.dot( x[i,:].T))))

  x_loss = np.array(x_loss)
  loss = np.sum(x_loss)
  return loss / len(y)

def cross_entropy(x, y, w):
  x_grad = []
  for i in range(0, len(y)):
    x_max = x[i,:].reshape(-1,1)
    y_max = y[i,:].reshape(-1,1)
    x_grad.append(sigmoid( -y_max.dot( w.T.dot( x_max)))*( -y_max.dot( x_max.T)))
  
  x_grad = np.array(x_grad)
  x_gradient = np.sum(x_grad, axis= 0) / len(y)
  return x_gradient 
```

```python
def pca_code(data):
    #raw_implementation
    var_per=.98
    data-=np.mean(data, axis=0)
    # data/=np.std(data, axis=0)
    cov_mat=np.cov(data, rowvar=False)
    evals, evecs = np.linalg.eigh(cov_mat)
    idx = np.argsort(evals)[::-1]
    evecs = evecs[:,idx]
    evals = evals[idx]
    variance_retained=np.cumsum(evals)/np.sum(evals)
    index=np.argmax(variance_retained>=var_per)
    evecs = evecs[:,:index+1]
    reduced_data=np.dot(evecs.T, data.T).T
    return evals
```
